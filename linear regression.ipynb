```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_data = pd.read_csv('salary.csv')
raw_data.head(59)
```

```python
raw_data['salaryBrutto'].replace(' ', np.nan, inplace=True)
filtered_data = raw_data[~pd.isnull(raw_data["salaryBrutto"])]
filtered_data.head(59)
```

```python
x_y = np.array(filtered_data)
x, y = x_y[:,0], x_y[:,1]
x, y = x.reshape(-1,1), y.reshape(-1, 1)
lin_regression = LinearRegression()
lin_regression.fit(x,y)
m = lin_regression.coef_
b = lin_regression.intercept_
predictions = lin_regression.predict(x)
print "y = [0]x + [1]".format(m, b)
```

```python
plt.scatter('workedYears','salaryBrutto', color='blue')
plt.plot([0,100],[b,m*100+b],'r')
plt.title('Linear Regression Example', fontsize = 20)
plt.xlabel('workedYears', fontsize = 15)
plt.ylabel('salaryBrutto', fontsize = 15)
plt.show()
```
